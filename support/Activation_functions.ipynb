```python
import numpy as np

def step(x):
  return np.where(x >= 0, 1, 0)

def sigmoid(x):
  return 1 / (1+np.exp(-x))

def tanh(x):
  return np.tanh(x)

def relu(x):
  return np.maximum(0, x)

def leaky_relu(x, alpha=0.15):
    return np.where(x >= 0, x, alpha * x)
```

```python
import matplotlib.pyplot as plt

# Create an array of x values
x = np.linspace(-5, 5, 100)  # Adjust the range and number of points as needed

# Create a 2x2 subplot grid
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

# Plot the activation functions in separate subplots
axes[0].plot(x, sigmoid(x))
axes[0].set_title('Sigmoid')

axes[2].plot(x, relu(x), label="ReLU")
axes[2].set_title('ReLU')

axes[1].plot(x, tanh(x))
axes[1].set_title('Tanh')


# Add labels and adjust spacing
for ax in axes.flat:
    ax.set_xlabel('x')
    ax.set_ylabel('Activation')
    ax.grid(True)

# Add overall title and adjust spacing
fig.suptitle('Activation Functions', fontsize=16)
plt.tight_layout()
plt.show()
```
